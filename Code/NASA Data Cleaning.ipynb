### Alameda County Land Plot 2021: Data Cleaning 

```python
#grabs the rounded x,y, and z values for the 2021 dataset

output1 = []

with open("filepath") as file1:
    for _ in range(20):
        next(file1)
    for line in file1:
        read = line.split()
        new_read = read[0:3]
        float_new_read = [float(value) for value in new_read]
        round_float = [round(value) for value in float_new_read]
        output1.append(round_float)
        
df1 = pd.DataFrame(output1, columns = ['X Coordinate #1', 'Y Coordinate #1', 'Z Coordinate #1'])
df1.head()
```

```python
#length of original 2021 LIDAR file
len(df1)
```

```python
#drops duplicate rows from the data (cells can still be represented by more than one z value currently)
df1_dropped_dup = df1.drop_duplicates()
df1_dropped_dup.rename(columns={'X Coordinate #1':'x','Y Coordinate #1':'y','Z Coordinate #1':'z1'},inplace = True)
df1_dropped_dup.head()
len(df1_dropped_dup)

#replace z coordinates with average of all columns with the same x and y, drop duplicates from new dataframe again 
pysqldf = lambda q: sqldf(q, globals())
avg = pysqldf('SELECT df1_dropped_dup.x, df1_dropped_dup.y, AVG(df1_dropped_dup.z1) AS z1 FROM df1_dropped_dup GROUP BY 1,2')
avg_dup_dropped = avg
avg_dup_dropped.head()
len(avg_dup_dropped)
```

### Alameda County Land Plot 2018: Data Cleaning

```python
#grabs the rounded x,y, and z values for the 2018 dataset
output3 = []

with open("filepath") as file3:
    for _ in range(20):
        next(file3)
    for line in file3:
        read = line.split()
        new_read = read[0:3]
        float_new_read = [float(value) for value in new_read]
        round_float = [round(value) for value in float_new_read]
        output3.append(round_float)
        
df3 = pd.DataFrame(output3, columns = ['X Coordinate #3', 'Y Coordinate #3', 'Z Coordinate #3'])
df3.head()
```

```python
#length of original 2018 LIDAR file
len(df3)
```

```python
#drops duplicate rows from the data (cells can still be represented by more than one z value currently)
df3_dropped_dup = df3.drop_duplicates()
df3_dropped_dup.rename(columns={'X Coordinate #3':'x','Y Coordinate #3':'y','Z Coordinate #3':'z3'},inplace = True)
df3_dropped_dup.head()
len(df3_dropped_dup)

#replace z coordinates with average of all columns with the same x and y, drop duplicates from new dataframe again 
pysqldf = lambda q: sqldf(q, globals())
avg3 = pysqldf('SELECT df3_dropped_dup.x, df3_dropped_dup.y, AVG(df3_dropped_dup.z3) AS z3 FROM df3_dropped_dup GROUP BY 1,2')
avg3_dup_dropped = avg3
avg3_dup_dropped.head()
len(avg3_dup_dropped)
```

```python
#Merge 2021, 2018, and 2010 tables together 
merge = pd.merge(avg_dup_dropped, avg2_dup_dropped, left_on=('x','y'), right_on=('x','y'), how='outer')
merge2 = pd.merge(merge, avg3_dup_dropped, left_on=('x','y'), right_on=('x','y'), how='outer')
merge2 = merge2.rename(columns={'x':'longitude','y':'latitude','z1':'height_2021','z2':'height_2010','z3':'height_2018'})
pysqldf('SELECT * FROM merge2 ORDER BY longitude,latitude')
```

```python
#Size of the total grid
(749*614)
```

```python
#Range of every potential longitude value
longitude = pd.DataFrame(range(561604, 562353), columns=['longitude'])
longitude.head()
```

```python
#Range of every potential latitude value
latitude = pd.DataFrame(range(4188653,4189267), columns=['latitude'])
latitude.head()
```

```python
#Every possible longitude and latitude combination (every possible cell, even those where there are no height records)
all_lon_lat = longitude.join(latitude, how='cross')
all_lon_lat
```

```python
#All cells that are not in the merge 
cells_not_in_table = all_lon_lat[~all_lon_lat.index.isin(merge2.index)]
cells_not_in_table
```

```python
#Percentage of total grid that doesn't have a height value reported among any of the 3 datasets
percent = len(cells_not_in_table) / len(all_lon_lat) * 100
round(percent, 2)
```

```python
#Combine cells we have height records for with the cells with no height values
merge3 = pd.merge(merge2, all_lon_lat, left_on=('longitude','latitude'), right_on=('longitude','latitude'), how='outer')
merge3.rename(columns={'x':'longitude','y':'latitude','z1':'height_2021','z2':'height_2010','z3':'height_2018'}, inplace=True)
complete_with_nulls = merge3.loc[:,['longitude','latitude','height_2021','height_2018','height_2010']]
complete_with_nulls
```

```python
#Used interpolation to fill null values
interpolated = complete_with_nulls.interpolate(method ='linear', limit_direction ='both')
interpolated
```
